```python
import pandas as pd
from sklearn.model_selection  import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/damodargupta/Desktop/ML/wine/wine.data')
df
```

```python
df.isnull().sum()
```

```python
# independent and dependent variables

X = df.iloc[: , :-1]
y = df.iloc[: , -1] # taking proline as our target variable
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.33 , random_state=42)
```

```python
X_train
```

```python
scalar = StandardScaler()
X_train_scaled = scalar.fit_transform(X_train)
X_test_scaled = scalar.transform(X_test)
```

```python
## PCA- Transformation
pca=PCA(n_components=12)
pca
```

```python
X_train_pca = pca.fit_transform(X_train_scaled)
X_train_pca
```

```python
explained_variance = pca.explained_variance_ratio_
explained_variance 
```

```python
# Plot the explained variance ratio
explained_variance = pca.explained_variance_ratio_
plt.bar(range(1, len(explained_variance) + 1), explained_variance)
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio for Each Principal Component')
plt.show()
```

### we get to know that the first three principal components capture almost 65 % of the variance of the whole dataset. so we select these 3 principal components. 

```python
# Scatter plot of the first three principal components
from mpl_toolkits.mplot3d import Axes3D

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_train_pca[:, 0], X_train_pca[:, 1], X_train_pca[:, 2], c='red', cmap='viridis')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('PCA Results (First Three Components)')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Specify the number of clusters
n_clusters = 3  # You can choose a different number

# Create a K-Means clustering object
kmeans = KMeans(n_clusters=n_clusters)

# Fit the K-Means model to the PCA-transformed data
kmeans.fit(X_train_pca)

# Get the cluster labels
cluster_labels = kmeans.labels_
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming X_pca contains the PCA-transformed data with 3 components
# Assuming cluster_labels contains the cluster assignments obtained from K-Means

# Create a 3D scatter plot with different colors for each cluster
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define colors for each cluster
cluster_colors = ['red', 'green', 'blue']  # You can add more colors for additional clusters

# Scatter plot each data point with a color corresponding to its cluster
for i in range(len(X_train_pca)):
    ax.scatter(X_train_pca[i, 0], X_train_pca[i, 1], X_train_pca[i, 2], c=cluster_colors[cluster_labels[i]], marker='o')

ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
ax.set_title('K-Means Clustering Results on PCA-transformed Data')
plt.show()
```

```python
# table showing performance metrics for clustering algorithm
import pandas as pd
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

# Assuming you have ground truth labels for comparison (not always available in unsupervised clustering)
# Replace 'true_labels' with the true labels if available
# If you don't have ground truth labels, consider using internal validation metrics

# Calculate clustering performance metrics
silhouette = silhouette_score(X_train_pca, cluster_labels)  # Silhouette Score
calinski_harabasz = calinski_harabasz_score(X_train_pca, cluster_labels)  # Calinski-Harabasz Score
davies_bouldin = davies_bouldin_score(X_train_pca, cluster_labels)  # Davies-Bouldin Index

# Create a dictionary with performance metrics
performance_metrics = {
    'Silhouette Score': [silhouette],
    'Calinski-Harabasz Score': [calinski_harabasz],
    'Davies-Bouldin Index': [davies_bouldin]
}

# Create a DataFrame to display the metrics
metrics_df = pd.DataFrame(performance_metrics)

# Display the table of performance metrics
print(metrics_df)
```
